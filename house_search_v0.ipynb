```python
import requests as req
import urllib.parse as parse
from bs4 import BeautifulSoup as soup
import json as json
import pandas as pandas


class Search:
    
     
    rent_min = 0
    rent_max = 0
    bedrooms = 0
    bathrooms = 0
    is_studio = ''
    is_pets = ''
    error_code = 0
    listing_DB = {}
    entry = {}

    apartment_baseURL = 'https://www.apartments.com/pittsburgh-pa/'
    hotpads_baseURL = 'https://hotpads.com/pittsburgh-pa/apartments-for-rent/'

    def __init__(self, rmin=None, rmax=None, pets=None, bedroom_count=None, bathroom_count=None, studio=None):

        if rmin == None:
            self.rent_min = 0
        else:
            self.rent_min = rmin


        if rmax == None:
            self.rent_max = 0
        else:
            self.rent_max = rmax

        if pets == None:
            self.is_pets = 'N'
        else:
            self.is_pets = pets

        if bedroom_count == None:
            self.bedrooms = 0
        else:
            self.bedrooms = bedroom_count

        if bathroom_count == None:
            self.bathrooms = 0
        else:
            self.bathrooms = bathroom_count

        if studio == None:
            self.is_studio = 'N'
        else:
            self.is_studio = studio

        self.error_code = 0
        self.listing_DB = {"entries": []}
        self.entry = {"entry_id": 0, "url": '', "image": '', "address_type": '', "street": '',
                      "phone": '', "zip": '', "prop_name": '', "prop_type": '', "prop_id": ''}

    def set_bedrooms(nrooms):
        self.bedrooms = nrooms
        pass

    def set_bathrooms(nrooms):
        self.bathrooms = nrooms
        pass

    def set_pets(aPets):
        self.is_pets = aPets
        pass

    def set_studio(aStudio):
        self.is_studio = aStudio
        pass

    def set_errorCodes(aError):
        self.error_code = aError
        pass

    def set_minimum_rent(rmin):
        self.rent_min = rmin
        pass

    def set_maximum_rent(rmax):
        self.rent_max = rmax
        pass

    def get_bedrooms(self):
        return (self.bedrooms)

    def get_bathrooms(self):
        return (self.bathrooms)

    def get_pets(self):
        return (self.is_pets)

    def get_studio(self):
        return (self.is_studio)

    def get_error_codes(self):
        return (self.error_code)

    def get_min_rent(self):
        return (self.rent_min)

    def get_max_rent(self):
        return (self.rent_max)

    def get_listing_DB(self):
        return (self.listing_DB)


    def get_bedRoom_param(self, source, bedroom_count):

        bedroom_query_str = ''


        if source == "Apartment":

            if bedroom_count == 0:
                bedroom_query_str = 'studios'

            else:
                bedroom_query_str = str(bedroom_count) + '-bedrooms'


        return bedroom_query_str


    def get_bathRoom_param(self, source, bathroom_count):

        bathRoom_query_str = ''

        if source == "Apartment":

            if bathroom_count == 0:
                bathRoom_query_str = '1-bathrooms'
            else:
                bathRoom_query_str = str(bathroom_count) + '-bathrooms'

        return bathRoom_query_str


    def get_pet_flag(self, pets_flag):
        if pets_flag == 'Y':
            pets = 'pet-friendly'
        else:
            pets = None
        return pets


    def get_URL(self, rent_cat, bedroom_cat, bathroom_cat, pets_cat, source):

        arr = []

        if rent_cat != None:
             arr.append(rent_cat)

        if bedroom_cat != None:
             arr.append(bedroom_cat)

        if bathroom_cat != None:
             arr.append(bathroom_cat)

        if pets != None:
             arr.append(pets)


        param_len = len(arr)

        z = param_len - 1


        if source == "Apartment":
            url = self.apartment_baseURL

        if param_len == 1:
             url = url + arr[0] + '/'
        else:

                for i in range(0,param_len):

                    if i == z:
                        url = url + arr[i] + '/'
                        print("3. in the last iter", i, z)
                    else:
                        url = url + arr[i] + '-'

        print("URL", url)

        return (url)


    def get_response(self, response):
        response = req.request("GET", response)
        return (response.status_code, response.text)


    def get_EntryID(self):
        return (len(self.listing_DB) + 1)


    def reset_resources(self):
        self.entry  = {"entry_id": 0, "url": '', "image": '', "address_type": '', "street": '',
                      "phone": '', "zip": '', "prop_name": '', "prop_type": '', "prop_id": ''}
        return self.entry


    def get_listings(self, response):

        soupy_o = soup(response, 'html.parser')

        resp = json.loads(soupy_o.find('script', type='application/ld+json').text)["about"]
        for props in resp:
            self.entry["entry_id"] = self.get_EntryID()
            self.entry["url"] = props["url"]
            self.entry["image"] = props["image"]
            self.entry["address_type"] = props["Address"]["@type"]
            self.entry["street"] = props["Address"]["streetAddress"]
            self.entry["zip"] = props["Address"]["postalCode"]
            self.entry["telephone"] = props["telephone"]
            self.entry["prop_type"] = props["@type"]
            self.entry["prop_id"] = props["@id"]
            self.listing_DB["entries"].append(self.entry)
            self.entry = self.reset_resources()

        return (self.listing_DB)


if __name__ == '__main__':
    rmin = 500
    rmax = 1000
    pets = 'Y'
    bedroom_count = 1
    bathroom_count = 1
    studio = 'N'

    search_req = Search(rmin, rmax, pets, bedroom_count, bathroom_count, studio)

    beds = search_req.get_bedRoom_param("Apartment", 1)
    baths = search_req.get_bathRoom_param("Apartment", 1)
    pets = search_req.get_pet_flag('Y')
    print("Pets", pets)

    url_str = search_req.get_URL("500-to-1000", beds, baths, pets, "Apartment")
    response = search_req.get_response(url_str)

    print("Listing", search_req.get_listings(response[1]))












```
